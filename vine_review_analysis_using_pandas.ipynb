```python
import pandas as pd
import os
```

```python
# Create the path and import the data
vine_review_df = os.path.join('Resources/CSV/vine_table.csv')
vine_df = pd.read_csv(vine_review_df)
```

```python
vine_df.info()
```

```python
vine_df['star_rating']= vine_df['star_rating'].astype('Int64')

vine_df.dtypes
```

```python
vine_df
```

```python
total_votes = vine_df.loc[vine_df['total_votes']>= 20]
total_votes
```

```python
total_votes_50 = total_votes.loc[(total_votes['helpful_votes']/total_votes['total_votes'])*100 >= 50]
total_votes_50
```

```python
paid_vine_program = total_votes_50.loc[total_votes_50['vine'] == 'Y']
print(len(paid_vine_program))
paid_vine_program
```

```python
unpaid_vine_program = total_votes_50.loc[total_votes_50['vine'] == 'N']
print(len(unpaid_vine_program))
unpaid_vine_program
```

```python
# Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for paid vine program
number_of_reviews = len(paid_vine_program)
number_of_5_star_reviews =len(paid_vine_program.loc[paid_vine_program['star_rating'] == 5]) 
percentage_of_5_star_review = (number_of_5_star_reviews/number_of_reviews) *100

print("Total number of reviews are: %f, The number of 5-star reviews are: %f and the percentage of 5-star reviews for paid vine program is: %f"
 %(number_of_reviews , number_of_5_star_reviews , percentage_of_5_star_review))
```

```python
# Determine the total number of reviews, the number of 5-star reviews, and the percentage of 5-star reviews for unpaid vine program
number_of_reviews_up = len(unpaid_vine_program)
number_of_5_star_reviews_up =len(unpaid_vine_program.loc[unpaid_vine_program['star_rating'] == 5]) 
percentage_of_5_star_review_up = (number_of_5_star_reviews_up/number_of_reviews_up) *100

print("Total number of reviews are: %f, The number of 5-star reviews are: %f and the percentage of 5-star reviews for unpaid vine program is: %f"
 %(number_of_reviews_up , number_of_5_star_reviews_up , percentage_of_5_star_review_up))
```
